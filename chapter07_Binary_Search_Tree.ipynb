```python
def contains(self, value):
    if self.root is None:  # None
        return False  #  False
    
    temp = self.root  # 
    
    while temp is not None:  # temp
        if value < temp.value:  # temp
            temp = temp.left  # 
        elif value > temp.value:  # 
            temp = temp.right  # 
        else:  # 
            return True  # True
    
    return False  # TrueFalse
```
